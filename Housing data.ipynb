### Data Preprocessing

**Data Cleaning or Cleansing**

```python
data.isnull().sum()
```

```python
# Avg. Area Number of Bedrooms have low correlation with house's price 
data.drop('Avg. Area Number of Rooms',axis=1,inplace=True)
```

```python
# dropping 'Adress' column
data.drop('Address',axis=1,inplace=True)
```

```python
data.head()
```

```python
X = data.drop('Price',axis=1)
```

```python
X.head()
```

```python
X.shape
```

```python
y=data['Price']
```

```python
y.shape
```

**Splitting the data to train and test sets**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Training the model 'Regression'

```python
from sklearn.linear_model import LinearRegression
```

```python
ln_reg=LinearRegression()
```

```python
ln_reg.fit(X_train,y_train)
```

```python
ln_reg.score(X_train,y_train)
```

```python
y_pred = ln_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error,r2_score
```

```python
print('R2_score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('RMSE',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
sns.distplot(y_test-y_pred) # plotting the error between the acual data and predicted data 
```

### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.10,random_state=0)
```

```python
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
```

```python
poly_reg=LinearRegression()
```

```python
poly_reg.fit(X_train_poly,y_train)
```

```python
y_pred=poly_reg.predict(X_test_poly)
```

```python
print('R2_score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('RMSE',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
sns.distplot(y_test-y_pred)
```
